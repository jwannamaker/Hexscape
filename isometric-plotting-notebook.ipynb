```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from src.geometry import hexagon

step = 60 # degrees
start = (180 - 60) / 2  # Orients polygon to have a flat ass
stop = start + 360 + 60
angles = np.arange(np.deg2rad(start), np.deg2rad(stop), np.deg2rad(step))
vertices = np.full((7, 3), 0.0)
for i, point in enumerate(vertices):
    point[0] = np.cos(angles[i])
    point[1] = np.sin(angles[i])
    point[2] = 0.0
print(vertices)
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.plot([v[0] for v in vertices], [v[1] for v in vertices])
```
